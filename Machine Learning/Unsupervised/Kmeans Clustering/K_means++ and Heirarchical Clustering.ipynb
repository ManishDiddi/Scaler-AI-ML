- This means, that the same data, with different initialization, will get different results (different clusters)
- The clusters formed in K-Means algorithm depend a lot on the way how initialize the centroids.

## K-Means ++ Algorithm

K-Means ++ gives us a better way to initialize the clusters.

1. Choose any existing data point as a centroid.
2. Then, while selecting the second centroid, choose another existing data point which is at farthest distance from the previous one.
3. Now, the third centroid should be at farthest distance from both the previous centroids.

</br>

<img src='https://drive.google.com/uc?id=1GLUR-93dgBACFhUfRYUaQ3wLGtyAuZzU' height='300' width='600'>

</br>

The only drawback is that it gives devastating results when we have an outlier present in the data.


**NOTE :**

## Hierarchical Clustering

Here, choosing the no. of clusters becomes slightly more intuitive as compared to KMeans/ KMeans ++.

There are two ways of performing Hierarchical Clustering
  
  1. Agglomerative (Popular and Widely used)
  2. Divisive

#### What is Agglomerative Clustering?

- The word agglomerative suggests combining things
- It is a **bottom-up** approach
- Agglomerative clustering starts with the assumption that every datapoint is a cluster
- Then, it groups the clusters which are closed to each other until there is only a single cluster left

<img src="https://drive.google.com/uc?export=view&id=1QFeq1V0dwWXwwM8WvCUbN3Z8Wrw14y3a">

#### What is Divisive Clustering?

- It is an complete opposite of agglomerative approach
- It is a **top-down** approach
- It starts with one big clusters that conatins all the datapoints.
- It then divides the points into different clusters till each data point is a cluster itself

#### Agglomerative clustering Case study

### What does the Agglomerative Algorithm look like?
Now, let's dive a little deep into Agglomerative approach


1. Assume each point is a cluster (n datapoints -> n clusters)
2. Compute Proximity Matrix  ($P_{n*n})$
3. Repeat until a single cluster is left:
  - Merge the closest clusters
  - Update the proximity matrix

- If we visualize this, this looks like a Tree, but there is other name which is often used in Data Mining terminology which is called **Dendogram**

### Q. How are we going to compute the similarity between groups of customers?

Let's say if we have two clusters then we can use - </br>

- **Single linkage** :
We take the minimum possible distance between any two points of the clusters.

- **Complete linkage**:
We take the farthest possible distance between any two points of the clusters.

<img src='https://drive.google.com/uc?id=1k4lvnmAZ3F8LkCrpa2bp8CU1r2HeNxPO' height='300' width='600'>

- **Average linkage:**
We take the average distance between all the points of one cluster to the other cluster.

- **Centroid linkage:**
We find out the centroid points of the two clusters and then calculate the distance between them.

<img src='https://drive.google.com/uc?id=1z9HwxACwQ0tBXqnQGVwJftZ9vm6NL5js' height='300' width='600'>

### **Ward method**

It sees that combining which two clusters is going to create minimum variance.

</br>

<img src='https://drive.google.com/uc?id=17S90_Z75dmj587c_MXpnwNJ-xrPVMs7l' height='300' width='600'>

</br>

  - First figure out the centroid of the two clusters.
  - Then calculate the variance of the data points in the cluster. </br> $var = \sum{d_i^2}$ </br>
    Where $d_i$ is the distance of each point with the centroid.
  - When we combine these two clusters, we again find a centroid and calculate the variance.
  - Now, the variance will definitely increase. </br> So we choose the two clusters to be merged in such a way that the increase in variance is minimal.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
import seaborn as sns

%matplotlib inline
````

````python
# reading the CSV file into pandas dataframe
custData = pd.read_csv("Cust_Spend_Data.csv")
custData.head(10)
````

````python
custDataAttr=custData.iloc[:,2:]
custDataAttr.head()
````

````python
custDataScaled=custDataAttr.apply(zscore)
custDataScaled.head(10)
````

Pairwise plot to understand realtionship between variables.

````python
# importing seaborn for statistical plots
sns.pairplot(custDataScaled, height=2,aspect=2 , diag_kind='kde')
````

````python
from scipy.cluster.hierarchy import dendrogram, linkage
````

##### **Dendogram using average linkage**

````python
Z = linkage(custDataScaled, metric='euclidean', method='average')
````

````python
Z.shape
````

````python
Z
````

````python
fig, ax = plt.subplots(figsize=(20, 12))
dendrogram(Z, labels=custDataScaled.index, ax=ax, color_threshold=2)
plt.xticks(rotation=90)
ax.set_ylabel('distance')
ax.set_xlabel('sample index')
ax.set_title("Dendogram using Average linkage")
````

##### **Dendogram using complete linkage**

````python
Z = linkage(custDataScaled, metric='euclidean', method='complete')
````

````python
fig, ax = plt.subplots(figsize=(20, 12))
dendrogram(Z, labels=custDataScaled.index, ax=ax, color_threshold=2)
plt.xticks(rotation=90)
ax.set_ylabel('distance')
ax.set_xlabel('sample index')
ax.set_title("Dendogram using Complete linkage")
````

##### **Dendogram using ward method**

````python
Z = linkage(custDataScaled, metric='euclidean', method='ward')
````

````python
fig, ax = plt.subplots(figsize=(20, 12))
dendrogram(Z, labels=custDataScaled.index, ax=ax, color_threshold=2)
plt.xticks(rotation=90)
ax.set_ylabel('distance')
ax.set_xlabel('sample index')
ax.set_title("Dendogram using Ward linkage")
````

#### **Calculating the number of clusters**

````python
fig, ax = plt.subplots(figsize=(20, 12))
dendrogram(Z, labels=custDataScaled.index, ax=ax, color_threshold=2)
plt.xticks(rotation=90)
plt.axhline(y=4, color='k', linestyle='--')
ax.set_ylabel('distance')
````

````python
from sklearn.cluster import AgglomerativeClustering
````

````python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean',  linkage='ward')
````

````python
model.fit(custDataScaled)
````

````python
custDataAttr['labels'] = model.labels_
custDataAttr.head(10)

#custDataAttr.groupby(["labels"]).count()
````

````python
# analysis of clusters formed
h_clusters_mthly_spend = 	pd.DataFrame(custDataAttr.groupby(["labels"]).Avg_Mthly_Spend.mean())
h_clusters_num_of_visits = 	pd.DataFrame(custDataAttr.groupby(["labels"]).No_Of_Visits.mean())
h_clusters_apparel_items = 	pd.DataFrame(custDataAttr.groupby(["labels"]).Apparel_Items.mean())
h_clusters_fnv_items = 	pd.DataFrame(custDataAttr.groupby(["labels"]).FnV_Items.mean())
h_clusters_staples_items = 	pd.DataFrame(custDataAttr.groupby(["labels"]).Staples_Items.mean())
````

````python
df = pd.concat([pd.Series([0,1,2]), h_clusters_mthly_spend, h_clusters_num_of_visits, h_clusters_apparel_items,h_clusters_fnv_items, h_clusters_staples_items ], axis=1)
df.columns = ["Cluster_id","Avg_Mthly_Spend","No_Of_Visits" ,"Apparel_Items", "FnV_Items", "Staples_Items"]
df.head()
````

````python
sns.barplot(x=df.Cluster_id, y=df.Avg_Mthly_Spend, hue=df.Cluster_id)
````

````python
sns.barplot(x=df.Cluster_id, y=df.No_Of_Visits, hue=df.Cluster_id)
````

````python
sns.barplot(x=df.Cluster_id, y=df.Apparel_Items, hue=df.Cluster_id)
````

````python
sns.barplot(x=df.Cluster_id, y=df.FnV_Items, hue=df.Cluster_id)
````

````python
sns.barplot(x=df.Cluster_id, y=df.Staples_Items, hue=df.Cluster_id)
````

Below are the insight that we can make out after analysing the above bar plots:
1. Customers in cluster 0 has the highest avg monthly spend, they frequently visit the store, they tend to prefer more apperal items and very lesser FnV and staple items.We can target these customer more apperal related ads.
2. Customers in cluster 1 has the second highest avg monthly spend, with highest number of visits to the store, they tend to buy more of FnV and staple items and almost no apperal items.We can target these customer more FnV and staple items related ads.

### Advantages and disatvantages of Aggloromative clustering

**Advantages of Hierarchical Clustering:**

- We do not need to specify the number of clusters required for the algorithm.
- Hierarchical clustering is easy to implement.
- The dendrogram produced is very useful in understanding the data.

**Disadvantages of Aggloromative clustering:**

- Not suitable for large datasets due to high time and space complexity.
  - The time required to run the algorithm increases exponentially as the number of data points increases, making it difficult to use for large datasets.
  - As data increases storage requirement would also increase because of similarity/proximity matrix that we are building.
