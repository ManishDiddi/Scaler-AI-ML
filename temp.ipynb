```python
def entropy(s):
    '''
    Calculates the entropy given list of target(binary) variables
    '''
    # Write your code here
    
    # Caclulate entropy
    entropy = 0
    p = np.mean(s)
    if p == 0 or p == 1:
        return 0.0
    entropy = p*np.log2(p)+(1-p)*np.log2(1-p)
    
    #Your code ends here
    
    return -entropy
```

```python
def information_gain(parent, left_child, right_child):
    
    '''
    Compute information gain given left_child target variables (list), right_child target variables(list) and their parent targets(list)
    '''
    
    info_gain=None
    # Write your code here
    parent_entropy = entropy(parent)
    left_child_entropy = entropy(left_child)
    right_child_entropy = entropy(right_child)
    n1,n2 = len(left_child),len(right_child)
    n = len(parent)
    weighted_children = (n1/n)*left_child_entropy + (n2/n)*right_child_entropy
    info_gain = parent_entropy - weighted_children
    
    #Your code ends here
    return info_gain
```

```python
y_true = np.array([1,0,2])
y_pred = np.array([1,0,1])
```

```python
classes = np.unique(y_true)
n_classes = len(np.unique(y_true))
n_classes
```

```python
confusion_matrix = np.zeros((n_classes,n_classes))
confusion_matrix
```

```python
label_to_idx = {c: i for i, c in enumerate(classes)}
label_to_idx
```

```python
for true,pred in zip(y_true,y_pred):
    confusion_matrix[true,pred] += 1
print(confusion_matrix)
```

```python
tp = np.diag(confusion_matrix).astype(float)
tp
```

```python
pred_sum = np.sum(confusion_matrix,axis=0).astype(float)
pred_sum
```

```python
true_sum = np.sum(confusion_matrix,axis=1).astype(float)
true_sum
```

```python
precision = np.divide(tp,pred_sum,where=pred_sum!=0,out=np.zeros_like(tp))
precision
```

```python
recall =np.divide(tp,true_sum,where=true_sum!=0,out=np.zeros_like(tp))
recall
```

```python
f1 = np.divide(2*precision*recall,(precision+recall),out=np.zeros_like(precision),where=(precision+recall)!=0)
f1
```
