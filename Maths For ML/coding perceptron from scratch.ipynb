```python
import numpy as np
```

```python
def perceptron(X, y, n_iter, eta):
  X = np.array(X)
  y = np.array(y)
  X_bias = np.c_[np.ones(X.shape[0]), X]

  w = np.ones((X_bias.shape[1]))

  for i in range(n_iter): # how many times for each point

    for xi, target in zip(X_bias, y): #point by point
      score = np.dot(w, xi) # predicted y
      if target==1 and score>0:
        continue
      elif target==1 and score<0:
        w = w+eta*target*xi
      elif target==-1 and score<0:
        continue
      else:
        w = w+eta*target*xi

  return w
```
